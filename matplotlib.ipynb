```python
#imporing matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matpoltlib inline
```

```python
#creating data
x = [100,222,150,300]
y = [1,2,3,4]
```

```python
#plotting the graph
plt.plot(y,x)
```

```python
#giving name to x ,y and title to the graph 
#specifing the color of the lines and it's size
plt.xlabel('number')
plt.ylabel('no')
plt.title('graph')
plt.plot(y,x,color='green',linewidth=20,linestyle='dotted')
```

```python
# easier way of specifing the color of the lines and it's size
# alpha is used to play with transparancy
plt.plot(y,x,'g+--',markersize=20,alpha=0.2)
```

```python
x = [100,222,150,300]
y = [1,2,3,4]
z=[6,3,4,34]
plt.plot(x,y)
plt.plot(x,z)
```

```python
x = [100,222,150,300]
y = [1,2,3,4]
z=[6,3,4,34]
plt.title('graph')
plt.xlabel('numbers')
plt.ylabel('nos')
plt.plot(x,y,label='avg')
plt.plot(x,z,label='min')
plt.legend()
plt.grid()
```

```python
#creating bar charts
y = ['goo','amz','msf','tsl']
x = [100,222,150,300]
x_pos=np.arange(len(y))
plt.bar(y,x)
```

```python
#creating a bar chat that has string column
y = ['goo','amz','msf','tsl']
x = [100,222,150,300]
z = [12,34,54,65]
x_pos=np.arange(len(y))
plt.bar(y,x)
plt.bar(y,z)
```

```python
#changing the size of the bar
y = ['goo','amz','msf','tsl']
x = [100,222,150,300]
z = [12,34,54,65]
x_pos=np.arange(len(y))
plt.bar(y,x,width=0.2)
plt.bar(y,z,width=0.2)
```

```python
#altering the position of the bar
y = ['goo','amz','msf','tsl']
x = [100,222,150,300]
z = [12,34,54,65]
x_pos=np.arange(len(y))
plt.bar(x_pos-0.2,x,width=0.2)
plt.bar(x_pos+0.2,z,width=0.2)
```

```python
#changing x axis to string again
y = ['goo','amz','msf','tsl']
x = [100,222,150,300]
z = [12,34,54,65]
x_pos=np.arange(len(y))
plt.xticks(x_pos,y)
plt.bar(x_pos-0.2,x,width=0.2)
plt.bar(x_pos+0.2,z,width=0.2)
```

```python
#creating horizontal bar plot
y = ['goo','amz','msf','tsl']
x = [100,222,150,300]
z = [12,34,54,65]
x_pos=np.arange(len(y))
plt.yticks(x_pos,y)
plt.barh(x_pos-0.2,x)
plt.barh(x_pos+0.2,z)
```

```python
#creating pie chart
vals = [1200,600,34,56,72]
ex_lables = ['rent','food','snacks','pen','juice']
plt.pie(vals)
```

```python
#to make the circle symmentrical
vals = [1200,600,34,56,72]
ex_lables = ['rent','food','snacks','pen','juice']
plt.axis('equal')
plt.pie(vals)
```

```python
#to hide unwanted details
vals = [1200,600,34,56,72]
ex_lables = ['rent','food','snacks','pen','juice']
plt.axis('equal')
plt.pie(vals)
plt.show()
```

```python
#to create lables
vals = [1200,600,34,56,72]
ex_lables = ['rent','food','snacks','pen','juice']
plt.axis('equal')
plt.pie(vals,labels=ex_lables)
plt.show()
```

```python
#to increase the circle size(radius)
#to make marking on the circle(autopct)
#to seprate pies from circle (explode)
vals = [1200,600,34,56,72]
ex_lables = ['rent','food','snacks','pen','juice']
plt.axis('equal')
plt.pie(vals,labels=ex_lables,radius=2,autopct='%0.2f%%',explode=[0,0.1,0,0,0])
plt.show()
```

```python
#creating a histogram
x=[21,35,13,45,87,64,100]
plt.hist(x)
```

```python
#limiting the no of bins to 3
x=[21,35,13,45,87,64,100]
plt.hist(x,bins=3,rwidth=0.9)
```

```python
#changing the orientation of the graph
x=[21,35,13,45,87,64,100]
plt.hist(x,bins=3,rwidth=0.9,orientation='horizontal')
```

```python
#printing two data at a time
x=[21,35,13,45,87,64,100]
y=[21,35,13,45,87,64,100]
plt.hist([x,y],bins=3,rwidth=0.9,color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
#changing the orientation of the histogram
x=[21,35,13,45,87,64,100]
y=[21,35,13,45,87,64,100]
plt.hist([x,y],bins=3,rwidth=0.9,color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend()
```
