```python
#setting date as index
df.set_index('day',inplace=True)
df
```

```python
#filling 0 in lace of nan and creating the copy of the table
df_new = df.fillna(0)
```

```python
#printing the old data frame
df
```

```python
#printing the new data frame
df_new
```

```python
#fill the nan values with the before cell's values
df_new = df.fillna(method='ffill')
df_new
```

```python
#fill the nan values with the next cell's values
df_new = df.fillna(method='bfill')
df_new
```

```python
#setting limit equals one lets to copy the content in before cell only once
df_new = df.fillna(method='bfill',limit=1)
df_new
```

```python
#it copies the adjecnt cell's values
df_new = df.fillna(method="bfill",axis="columns")
df_new
```

```python
#fillinng o in the temp and wind_speed column and no event in event column in nan value cells
df_new = df.fillna({
    'temprature':0,
    'windspeed':0,
    'event':'no event'
})
df_new
```

```python
#prints the avg of before and after cell
df.interpolate()
```

```python
#time specifies the date and the value is taken avg accordingly
df.interpolate(method="time")
```

```python
#drops all the nan value tables
df.dropna()
```

```python
#drop thr row with all nan values
df.dropna(how="all")
```

```python
#drops rows with more than one nan values
df.dropna(thresh=1)
```
