handling missing data (replace function)

```python
#reading a csv file
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\G.M.BHOOPESH\Desktop\ds\study materials\pandas\weatherdata.csv")
df
```

```python
#using replace function replace -99999 with 0
df_new = df.replace(-99999,0)
df_new
```

```python
#replacing by creating dictionary
df_new1 = df.replace({
    "temperature": -99999,
    "windspeed": -99999,
    "event":'0'
},np.NaN)
df_new1
```

```python
#np.NaN is used to replace NaN in the specified column note:numpy must be imported first
df_new2 = df.replace([-99999,'0'],np.NaN)
df_new2
```

```python
#regex used to replace all aplha values to NaN
df_new3 = df.replace('[A-Za-z]',np.NaN,regex=True)
df_new3 
```

```python
#creating a dataframe manually
df_new4 = pd.DataFrame({
    'name':["lalith",'dharan','charan','karan','kiran'],
    'score':['pass',"topper",'fail','justpass','pass']
})
df_new4
```
